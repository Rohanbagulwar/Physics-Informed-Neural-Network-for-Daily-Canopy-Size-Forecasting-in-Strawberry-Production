```python
import pandas as pd

# File paths
file1 = r"F:\strwaberry images\canopy_volume_excels\top5_preprocessed_images.xlsx"   # Contains: image_name, depth_volume
file2 = r"F:\strwaberry images\Back_row_excel_dataset_copy\final_dataset_back_setup_top_plant5.xlsx"   # Contains: image_name
# file2 = ""   # Contains: image_name, depth_volume (to be updated)
# Read both Excel files
df1 = pd.read_excel(file1)
df2 = pd.read_excel(file2)

# Clean column names (optional but safe)
df1.columns = df1.columns.str.strip()
df2.columns = df2.columns.str.strip()

# Create mapping dictionary from df1
# Drop duplicates if any
mapping = df1.set_index("image_name")["depth_volume"].to_dict()

# Map depth_volume to df2 based on image_name
df2["depth_volume"] = df2["Image_name"].map(mapping)

# Ensure blanks remain blank (not NaN)
df2["depth_volume"] = df2["depth_volume"].where(pd.notna(df2["depth_volume"]), "")



# print(f"Mapping completed. File saved as {output_file}")
```

```python
import pandas as pd
import os

# Folder path
folder_path = "Final_Dataset_with_canopy"

# Get all Excel files
files = [f for f in os.listdir(folder_path) if f.endswith('.xlsx')]

# Read and combine
df_list = []
for file in files:
    file_path = os.path.join(folder_path, file)
    df = pd.read_excel(file_path)
    df_list.append(df)

# Merge all
final_df = pd.concat(df_list, ignore_index=True)

# Save merged file
final_df.to_excel("merged_final_data.xlsx", index=False)

print("Merged successfully!")

```

```python
import os
import shutil

# Source and destination folders
source_folder = "top2_preprocessed_images_lighting_removal"
destination_folder = "top2_new_preprocessed_images_lighting_removal"

# Create destination folder if not exists
os.makedirs(destination_folder, exist_ok=True)

# Loop through files
for filename in os.listdir(source_folder):
    if filename.startswith("Down2_"):
        
        # New filename
        new_filename = filename.replace("Down2_", "Top2_", 1)
        
        # Full paths
        src_path = os.path.join(source_folder, filename)
        dest_path = os.path.join(destination_folder, new_filename)
        
        # Copy and rename
        shutil.copy(src_path, dest_path)

print("Files renamed and saved successfully!")
```
